```python
def get_simstock_result_abl(data_name):
    us = data_name #nasdaq_nasdaq_2022
    us = us[["Date","Stock_","Label"]]
    df_grouped = us.groupby('Stock_')['Label'].apply(list).reset_index()
    df_expanded = pd.DataFrame(df_grouped['Label'].tolist(), index=df_grouped['Stock_']).transpose()
    df_expanded = df_expanded.iloc[:us[us["Stock_"] == us["Stock_"].unique()[0]].shape[0], :]
    df_expanded.index = us[us["Stock_"] == us["Stock_"].unique()[0]].set_index("Date").index
    df_expanded = df_expanded.dropna(axis = 1)
    return df_expanded
```

```python
def calculate_diagonal_perform_abl(corr1, data_01, data_03, data_05, data_07, data_09, gt_corr, gt_dtw, num_stocks):
    all_stock_list = list(corr1.columns)
    result_corr_list = []
    result_dtw_list = []
    corr1.index = corr1.columns
    
    for i in tqdm(all_stock_list, desc="Processing stocks"):
        corr1_list_ = list(corr1[i].nlargest(num_stocks).index)
        data_01_list = list(data_01[i].nlargest(num_stocks).index)
        data_03_list = list(data_03[i].nlargest(num_stocks).index)
        data_05_list = list(data_05[i].nlargest(num_stocks).index)
        data_07_list = list(data_07[i].nlargest(num_stocks).index)
        data_09_list = list(data_09[i].nlargest(num_stocks).index)
        
        result_corr = {
            'symbol': i,
            'data_01_mean': round(gt_corr.loc[gt_corr.columns.intersection(data_01_list)][i].values[1:].mean(), 3),
            'data_03_mean': round(gt_corr.loc[gt_corr.columns.intersection(data_03_list)][i].values[1:].mean(), 3),
            'data_05_mean': round(gt_corr.loc[gt_corr.columns.intersection(data_05_list)][i].values[1:].mean(), 3),
            'data_07_mean': round(gt_corr.loc[gt_corr.columns.intersection(data_07_list)][i].values[1:].mean(), 3),
            'data_09_mean': round(gt_corr.loc[gt_corr.columns.intersection(data_09_list)][i].values[1:].mean(), 3),


        }
        
        result_corr_list.append(result_corr)
        result_dtw = {
            'symbol': i,
            'data_01_mean': round(gt_dtw.loc[gt_dtw.columns.intersection(data_01_list)][i].values[1:].mean(), 3),
            'data_03_mean': round(gt_dtw.loc[gt_dtw.columns.intersection(data_03_list)][i].values[1:].mean(), 3),
            'data_05_mean': round(gt_dtw.loc[gt_dtw.columns.intersection(data_05_list)][i].values[1:].mean(), 3),
            'data_07_mean': round(gt_dtw.loc[gt_dtw.columns.intersection(data_07_list)][i].values[1:].mean(), 3),
            'data_09_mean': round(gt_dtw.loc[gt_dtw.columns.intersection(data_09_list)][i].values[1:].mean(), 3),


            }
        result_dtw_list.append(result_dtw)
    result_corr = pd.DataFrame(result_corr_list)
    result_dtw = pd.DataFrame(result_dtw_list)        
    result_corr["top_k"] = num_stocks-1
    result_dtw["top_k"] = num_stocks-1        
    return result_corr, result_dtw
        
        
```
