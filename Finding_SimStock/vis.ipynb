```python
import pandas as pd
import os
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def get_mean_std(data):
    data = data.fillna(0.0) # for Top 1 std values
    data.index=  ["TOP@1","TOP@3","TOP@5","TOP@7","TOP@9"] 
    mean_ = data.iloc[:, :5].T
    std_ = data.iloc[:, 5:].T

    mean_.index  = ["Corr1","Corr2","Peer","SimStock","TS2VEC"] 
    std_.index  = ["Corr1","Corr2","Peer","SimStock","TS2VEC"] 
    return mean_, std_

def get_mean_std_non_off(data):
    data = data.fillna(0.0) # for Top 1 std values
    data.index=  ["TOP@1","TOP@3","TOP@5","TOP@7","TOP@9"] 
    mean_ = data.iloc[:, :4].T
    std_ = data.iloc[:, 4:].T

    mean_.index  = ["Corr1","Corr2","SimStock","TS2VEC"] 
    std_.index  = ["Corr1","Corr2","SimStock","TS2VEC"] 
    return mean_, std_

import matplotlib.font_manager as fm
font_path = './font/Times New Roman.ttf'  
fm.fontManager.addfont(font_path)

import matplotlib.pyplot as plt
plt.rc('font', family='Times New Roman')
```
